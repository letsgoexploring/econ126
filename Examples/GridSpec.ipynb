Example from:

https://matplotlib.org/3.1.0/tutorials/intermediate/gridspec.html

```python
import matplotlib.pyplot as plt
```

```python
fig3 = plt.figure(constrained_layout=True)
gs = fig3.add_gridspec(3, 3)
ax1 = fig3.add_subplot(gs[0, :])
ax1.set_title('gs[0, :]')
ax2 = fig3.add_subplot(gs[1, :-1])
ax2.set_title('gs[1, :-1]')
ax3 = fig3.add_subplot(gs[1:, -1])
ax3.set_title('gs[1:, -1]')
ax4 = fig3.add_subplot(gs[-1, 0])
ax4.set_title('gs[-1, 0]')
ax5 = fig3.add_subplot(gs[-1, -2])
ax5.set_title('gs[-1, -2]')
```
