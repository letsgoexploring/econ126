## Exercise: Strings

Follow the instructions in the following cell. *Be sure to spell all variable names correctly.*

```python
# Create a variable called `city` that equals "irVinE". Print the value of the variable `city`.
# BEGIN SOLUTION
city = 'irVinE'
print(city)
# END SOLUTION

# Use the string method `title()` on the variable `city` to create a variable called `city_title`
#    that has its first character capitalized. Print the value of the variable `city_title`.
# BEGIN SOLUTION
city_title = city.title()
print(city_title)
# END SOLUTION

# Use the string method `upper()` on the variable `city` to create a variable called `city_upper`
#    that has all characters capitalized. Print the value of the variable `city_upper`.
# BEGIN SOLUTION
city_upper = city.upper()
print(city_upper)
# END SOLUTION

# Use the string method `swapcase()` on the variable `city` to create a variable called `city_swapped`
#    that reverses the case of each character in `the variable city`. Print the value of the variable 
#    `city_swapped`.
# BEGIN SOLUTION
city_swapped = city.swapcase()
print(city_swapped)
# END SOLUTION

# Use the string method `replace()` on the variable `city` to create a variable called `city_replaced`
#    that replaces all `i`s in the variable `city` by `a`s. Print the value of the variable
#    `city_replaced`.
# BEGIN SOLUTION
city_replaced = city.replace('i','a')
print(city_replaced)
# END SOLUTION

# Create a new variable called `state` equal to `CA`. Use `city` and `state` (and the `title()` method) 
#     to print `Irvine, CA`.
# BEGIN SOLUTION
state = 'CA'
print(city.title()+', '+state)
# END SOLUTION

# Use the string methods `join()` and `title()` to combine the variables `city` and `state` into 
#    `Irvine, CA` and print `Irvine, CA`.
# BEGIN SOLUTION
print(', '.join([city.title(),state]))
# END SOLUTION
```
