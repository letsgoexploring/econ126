## Exercise: Strings

Follow the instructions in the following cell. *Be sure to spell all variable names correctly.*

```python
# Create variable `city` and print
# BEGIN SOLUTION
city = 'irVinE'
print(city)
# END SOLUTION

# Create variable `city_title` and print
# BEGIN SOLUTION
city_title = city.title()
print(city_title)
# END SOLUTION

# Create variable `city_upper` and print
# BEGIN SOLUTION
city_upper = city.upper()
print(city_upper)
# END SOLUTION

# Create variable `city_swapped` and print
# BEGIN SOLUTION
city_swapped = city.swapcase()
print(city_swapped)
# END SOLUTION

# Create variable `city_replaced` and print
# BEGIN SOLUTION
city_replaced = city.replace('i','A')
print(city_replaced)
# END SOLUTION

# Create a new variable called `state` equal to `CA`. Use `city` and `state` (and the `title` method) 
#     to print `Irvine, CA`.
# BEGIN SOLUTION
state = 'CA'
print(city.title()+', '+state)
# END SOLUTION

# Use the string methods `join` and `title` to combine `city` and `state` into `Irvine, CA` and print
#    `Irvine, CA`.
# BEGIN SOLUTION
print(', '.join([city.title(),state]))
# END SOLUTION
```

```python
# HIDDEN
assert city == 'irVinE'
```

```python
# HIDDEN
assert city_title == city.title()
```

```python
# HIDDEN
assert city_upper == city.upper()
```

```python
# HIDDEN
assert city_swapped == city.swapcase()
```

```python
# HIDDEN
assert city_replaced == city.replace('i','A')
```
