```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from fredpy import series, window_equalize
%matplotlib inline
```

```python
# Download nominal GDP, nominal personal consumption expenditures, nominal
# gross private domestic investment, the GDP deflator, and an index of hours 
# worked in the nonfarm business sector produced by the BLS. All data should
# be quarterly.

gdp = series('GDP')
cons = series('PCEC')
invest = series('GPDI')
hours = series('HOANBS')
defl = series('GDPDEF')
```

```python
# Make sure that all of the downloaded series have the same data ranges

gdp,cons,invest,hours,defl = window_equalize([gdp,cons,invest,hours,defl])

print(gdp.daterange)
print(cons.daterange)
print(invest.daterange)
print(hours.daterange)
print(defl.daterange)
```

```python
# Compute real GDP, real consumption, real investment
gdp.data = gdp.data/defl.data*100
cons.data = cons.data/defl.data*100
invest.data = invest.data/defl.data*100
```

```python
# Compute and plot log real GDP, log consumption, log investment, log hours

gdp_log = gdp.log()
cons_log = cons.log()
invest_log = invest.log()
hours_log = hours.log()
```

```python
# Plot real GDP, real consumption, real investment, and hours in a 2x2 grid. Include recession bars

fig = plt.figure(figsize=(12,8))

ax1 = fig.add_subplot(2,2,1)
ax1.plot_date(gdp.datetimes,gdp.data,'-',lw=3,alpha = 0.7)
ax1.grid()
ax1.set_title(gdp.title)
ax1.set_ylabel(gdp.units)
gdp.recessions()


ax2 = fig.add_subplot(2,2,2)
ax2.plot_date(cons.datetimes,cons.data,'-',lw=3,alpha = 0.7)
ax2.grid()
ax2.set_title(cons.title)
ax2.set_ylabel(cons.units)
cons.recessions()

ax3 = fig.add_subplot(2,2,3)
ax3.plot_date(invest.datetimes,invest.data,'-',lw=3,alpha = 0.7)
ax3.grid()
ax3.set_title(invest.title)
ax3.set_ylabel(invest.units)
invest.recessions()

ax4 = fig.add_subplot(2,2,4)
ax4.plot_date(hours.datetimes,hours.data,'-',lw=3,alpha = 0.7)
ax4.grid()
ax4.set_title(hours.title)
ax4.set_ylabel(hours.units)
hours.recessions()

fig.tight_layout()
```
